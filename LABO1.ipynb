## Problema 2

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def reconstruir_imagen_color(img_azul_path, img_verde_path, img_rojo_path, output_path, titulo):
    # Se cargan las imgenes en escala de grises
    img_azul = cv2.imread(img_azul_path, cv2.IMREAD_GRAYSCALE)
    img_verde = cv2.imread(img_verde_path, cv2.IMREAD_GRAYSCALE)
    img_rojo = cv2.imread(img_rojo_path, cv2.IMREAD_GRAYSCALE)

    # Se verifica si las imgenes se cargaron correctamente
    if img_azul is None:
        print(f"Error: No se pudo cargar la imagen azul desde {img_azul_path}")
        return
    if img_verde is None:
        print(f"Error: No se pudo cargar la imagen verde desde {img_verde_path}")
        return
    if img_rojo is None:
        print(f"Error: No se pudo cargar la imagen roja desde {img_rojo_path}")
        return

    # Se crea una imagen vaca con tres canales (color)
    img_color = np.zeros((img_azul.shape[0], img_azul.shape[1], 3), dtype=np.uint8)

    # Se asigna cada imagen de escala de grises a su respectivo canal de color
    img_color[:, :, 0] = img_azul  # Canal azul
    img_color[:, :, 1] = img_verde  # Canal verde
    img_color[:, :, 2] = img_rojo  # Canal rojo

    # Se guarda la imagen resultante
    cv2.imwrite(output_path, img_color)

    # Se muestra la imagen resultante en el notebook
    plt.figure(figsize=(8, 6))
    plt.imshow(cv2.cvtColor(img_color, cv2.COLOR_BGR2RGB))
    plt.title(titulo)
    plt.axis('off')
    plt.show()

# Rutas de las imgenes y los directorios de salida
img_paths = [
    (r'C:\Users\walte\Documents\Text Mining\imagen1\imagen1_salida_gray_azul.jpg',
     r'C:\Users\walte\Documents\Text Mining\imagen1\imagen1_salida_gray_verde.jpg',
     r'C:\Users\walte\Documents\Text Mining\imagen1\imagen1_salida_gray_rojo.jpg',
     r'C:\Users\walte\Documents\Text Mining\imagen_color_reconstruida.jpg',
     'Imagen 1 Color Reconstruida'),
    (r'C:\Users\walte\Documents\Text Mining\imagen2\imagen2_salida_gray_azul.jpg',
     r'C:\Users\walte\Documents\Text Mining\imagen2\imagen2_salida_gray_verde.jpg',
     r'C:\Users\walte\Documents\Text Mining\imagen2\imagen2_salida_gray_rojo.jpg',
     r'C:\Users\walte\Documents\Text Mining\imagen2_color_reconstruida.jpg',
     'Imagen 2 Color Reconstruida'),
    (r'C:\Users\walte\Documents\Text Mining\perro\perro_salida_gray_azul.jpg',
     r'C:\Users\walte\Documents\Text Mining\perro\perro_salida_gray_verde.jpg',
     r'C:\Users\walte\Documents\Text Mining\perro\perro_salida_gray_rojo.jpg',
     r'C:\Users\walte\Documents\Text Mining\perro_color_reconstruida.jpg',
     'Imagen Perro Color Reconstruida')
]

# Llamar a la funcin para cada conjunto de imgenes
for img_azul_path, img_verde_path, img_rojo_path, output_path, titulo in img_paths:
    reconstruir_imagen_color(img_azul_path, img_verde_path, img_rojo_path, output_path, titulo)
```

## Problema 4:
Cree una funcion que dada una imagen, muestre el histograma de cada canal de color y el de
escala de grises (utilice un promedio atitmetico para su escala de grises, no puede usar funciones
de opencv), sus histogramas deben incluir una linea vertical la cual muestre el valor de la media
de la distribucion

```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

def mostrar_histogramas_con_media(img_path):
    # Se carga la imagen a color usando PIL
    img_color = Image.open(img_path)
    img_color = np.array(img_color)
    
    # Se convierte la imagen a escala de grises manualmente
    img_gray = np.dot(img_color[...,:3], [0.2989, 0.5870, 0.1140])
    
    # Se separa los canales de color
    azul = img_color[:, :, 2]
    verde = img_color[:, :, 1]
    rojo = img_color[:, :, 0]
    
    # Se crea una lista de los canales y sus nombres
    canales = [azul, verde, rojo, img_gray]
    nombres = ['Azul', 'Verde', 'Rojo', 'Gris']
    colores = ['b', 'g', 'r', 'k']
    
    # Se configura la figura para los histogramas
    plt.figure(figsize=(12, 8))
    
    for i, canal in enumerate(canales):
        plt.subplot(2, 2, i + 1)
        plt.title(f'Histograma del canal {nombres[i]}')
        plt.xlabel('Intensidad')
        plt.ylabel('Frecuencia')
        
        # Se calcula el histograma manualmente
        hist, bins = np.histogram(canal.flatten(), bins=256, range=[0,256])
        
        # Se calcula la media
        media = np.mean(canal)
        
        # Se grafica el histograma
        plt.plot(hist, color=colores[i])
        
        # Se anada una lnea vertical en el valor de la media
        plt.axvline(media, color='k', linestyle='dashed', linewidth=1)
        plt.text(media + 5, max(hist) * 0.9, f'Media: {media:.2f}', color='k')
    
    # Se muestra los histogramas
    plt.tight_layout()
    plt.show()

# Ruta de la imagen
img_path = r'C:\Users\walte\Documents\Text Mining\TIKAL.jpg'

# Llamar a la funcin
mostrar_histogramas_con_media(img_path)
```

## Enfoque de Escala de Grises Ponderado

El enfoque de escala de grises ponderado consiste en convertir una imagen a color (RGB) en una imagen en escala de grises asignando diferentes pesos a cada uno de los canales de color (rojo, verde, azul). Esto se hace porque cada color contribuye de manera diferente a la percepcin de brillo en una imagen.

Frmula Comn
Una frmula comnmente usada para convertir una imagen RGB a escala de grises ponderada es:

Gris=0.2989Rojo+0.5870Verde+0.1140Azul

Estos pesos se basan en la sensibilidad del ojo humano a estos colores. Sin embargo, estos pesos pueden ajustarse segn diferentes criterios


## Implementacin

```python

from PIL import Image
import matplotlib.pyplot as plt

def escala_grises_ponderada(img_path, output_path):
    # Se carga la imagen a color usando PIL
    img_color = Image.open(img_path)
    img_color = np.array(img_color)
    
    # Se convierte la imagen a escala de grises ponderada
    img_gray = np.dot(img_color[...,:3], [0.2989, 0.5870, 0.1140])
    
    # Se convierte el resultado a imagen PIL para guardarla
    img_gray_pil = Image.fromarray(img_gray.astype('uint8'), 'L')
    img_gray_pil.save(output_path)
    
    # Se muestra la imagen original y la imagen en escala de grises
    plt.figure(figsize=(12, 6))
    
    plt.subplot(1, 2, 1)
    plt.title('Imagen Original')
    plt.imshow(img_color)
    
    plt.subplot(1, 2, 2)
    plt.title('Imagen en Escala de Grises Ponderada')
    plt.imshow(img_gray, cmap='gray')
    
    plt.show()

# Rutas de la imagen de entrada y la imagen de salida
img_path = r'C:\Users\walte\Documents\Text Mining\IZABAL.jpg'
output_path = r'C:\Users\walte\Documents\Text Mining\IZABAL_grayscale_ponderada.jpg'

# Llamar a la funcin
escala_grises_ponderada(img_path, output_path)
```

Investigue brevemente en que consiste el espacio de color HSV y como se mapean colores a dicho

RGB es un modelo de color basado en los colores primarios aditivos: rojo, verde, azul.

El modelo de color HSV se basa en un sistema de color en el que el espacio de color se representa mediante un nico cono: tono, saturacin y valor.

La funcin Conversin de modelo de color se puede utilizar en un dataset de mosaico.

Esta funcin solo admite entradas de 3 bandas.
A continuacin, se explica cada componente del espacio de color HSV:

##### Hue (Tono):

Representa el color puro y se mide en grados de 0 a 360 en el crculo cromtico.
0 es rojo, 120 es verde, 240 es azul, y as sucesivamente.
El tono define el tipo de color que percibimos, como rojo, azul, verde, etc.

##### Saturation (Saturacin):

Representa la intensidad o pureza del color.
Se mide en un rango de 0 a 1 (o 0% a 100%).
Un valor de 0 significa que el color es un tono de gris (sin color), mientras que un valor de 1 significa un color completamente puro sin mezclas de blanco o negro.

##### Value (Valor o Brillo):

Representa la luminosidad o brillo del color.
Se mide en un rango de 0 a 1 (o 0% a 100%).
Un valor de 0 es completamente negro (sin luz), mientras que un valor de 1 es el color ms brillante posible.
